```python
#  Basic feature extraction using text data
import pandas as pd
from six.moves import urllib
array = []
url = "https://raw.githubusercontent.com/unt-iialab/info5731_spring2021/main/class_exercises/01-05-1%20%20Adams%20v%20Tanner.txt"
x = urllib.request.urlopen(url)
for f in x:
  d = f.decode("utf-8").replace('\r\n', '')
  if d:
    array.append(d)

df = pd.DataFrame (array, columns = ['Raw Text'])    
print(df)



```

```python
#total sentences
df['No of sentences'] = df['Raw Text'].apply(lambda x: len(str(x).split(".")))
df
```

```python
#total words
df['No of Words'] = df['Raw Text'].apply(lambda x: len(str(x).split(" ")))
df
```

```python
#total characters
df['No of Characters'] = df['Raw Text'].str.len()
df
```

```python
#average word length
def avg(sent):
  x = sent.split()
  if len(x) != 0:
    return(sum(len(word) for word in x)/len(x))
  else:
    return None
df['Avg Word Length'] = df['Raw Text'].apply(lambda x: avg(x))
df
```

```python
#number of stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
df['No of Stopwords'] = df['Raw Text'].apply(lambda x: len([x for x in x.split() if x in stop]))
df
```

```python
#number of special characters
def char(sent):
  x = 0
  for y in range(len(sent)):
    if not((sent[y].isalpha()) and (sent[y].isdigit())):
        x = x + 1
  return x
df['No of special characters'] = df['Raw Text'].apply(lambda x: char(x))
df
```

```python
#number of numerics
df['No of numerics'] = df['Raw Text'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
df
```

```python
#Number of uppercase words
df['No of upper case words'] = df['Raw Text'].apply(lambda x: len([x for x in x.split() if x.isupper()]))
df
```

```python
#Basic Text Pre-processing of text data

#Lower casing
df['Lower Case'] = df['Raw Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```

```python
#Punctuation removal
df['Punctuation removal'] = df['Lower Case'].str.replace('[^\w\s]','')
df
```

```python
#Stopwords removal
import nltk
from nltk.corpus import stopwords
remove = stopwords.words('english')
df['Stopwords Removal'] = df['Punctuation removal'].apply(lambda x: " ".join(x for x in x.split() if x not in remove))
df
```

```python
#Frequent words removal
x = pd.Series(' '.join(df['Stopwords Removal']).split()).value_counts()[:10]
print(x)
y = list(x.index)
df[' Removed common words'] = df['Stopwords Removal'].apply(lambda x: " ".join(x for x in x.split() if x not in y))
df
```

```python
#Rare words removal
x = pd.Series(' '.join(df['Stopwords Removal']).split()).value_counts()[-10:]
print(x)
y = list(x.index)
df[' Removed rare words'] = df['After common words removal'].apply(lambda x: " ".join(x for x in x.split() if x not in y))
df
```

```python
#Spelling correction
from textblob import TextBlob
df['Spelling Correction'] = df['After Rare Words removal'].apply(lambda x: str(TextBlob(x).correct()))
df
```

```python
#Tokenization
import nltk
nltk.download('punkt')
df['Tokenization'] = df['Spelling Correction'].apply(lambda x: TextBlob(x).words)
df
```

```python
#Stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
df[' Stemming'] = df['Tokenization'].apply(lambda x: " ".join([st.stem(word) for word in x]))
df
```

```python
#Lemmatization
from textblob import Word
import nltk
nltk.download('wordnet')
df['Lemmatization'] = df[' Stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```

```python
#saving it to csv
df.to_csv('cleaned_data.csv',index=False)
```

```python
#Advance Text Processing

#Calculating Term frequencies
x = (df['Lemmatization']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
x.columns = ['words','term_fre']
x
```
